```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# number of samples:
n = 100 

# uniform rand dist signal
y_uniform = np.random.uniform(low =-0.5, high=0.5, size=n)
meanVal_uni = np.mean(y_uniform);

# norm (gauss) rand dist signal
mu, sigma = 0,.5 # set mean and std 
y_normal = np.random.normal(mu, sigma,n)
meanVal_norm = np.mean(y_normal);

plt.figure(figsize=(8,6));
plt.plot(range(n),y_uniform, label = 'uniform')
plt.plot(range(n),y_normal, label = 'normal')

plt.title('Mean value of uniform signal = '+ str(meanVal_uni))
plt.show
plt.xlabel('n (Sample #)')
plt.ylabel('y[n]')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
